## Advanced Data Manipulation in Python
### Selecting Columns: filter

```python
# columns `Team_type`, `human_work`, and `work_per_human`
(gh_teams
 .filter(["Team_type", "human_work", "work_per_human"])
) 
```

```python
# columns that end in the letter `t`
(gh_teams
 .filter(regex = "t$")
)
```

### Sorting Values

```python
# sorting descending on columns `Team_size_class`, `human_work`, `work_per_human`
(gh_teams
 .sort_values(by = [ "Team_size_class", "human_work", "work_per_human"], ascending = False)
)
```

### Selecting Values: query

```python
# select `human-bot` teams that have a `bot_members_count` value greater than and equal to 3
(gh_teams
 .query('Team_type == "human-bot" & bot_members_count >= 3')
)
```

```python
# 'human` teams that are `Large` and have a `human_gini` value greater than and equal to 0.75
(gh_teams
 .query('Team_type == "human" & Team_size_class == "Large" & human_gini >= 0.75')
)
```

### Finding/Counting Data

```python
# teams in the `Small` or `Large` category - 147 teams
(gh_teams
 .query('Team_size_class == "Large" | Team_size_class == "Small"')
)
```

There are 428 teams in the Small or Large categories

```python
# teams  in the `Small` or `Large` category with a `human_gini` value less than and equal to 0.25

(gh_teams
 .query('(Team_size_class == "Large" | Team_size_class == "Small") &  human_gini >= 0.25')
)
```

There are 342 teams in the Small or Large categories with a `human_gini` value less than and equal to 0.25.

```python
# `human` teams in the `Medium` category
(gh_teams
 .query('Team_type == "human" & Team_size_class == "Medium" ')
)
```

There are 96 `human` teams in the `Medium` category.
